```python
#Перевод единиц измерения
import pandas as pd
import numpy as np

def convert_to_mg_ml(row):
  value = row['Standard Value']
  units = row['Standard Units']
  molecular_weight = row['Molecular Weight']

  if pd.isnull(value) or pd.isnull(units) or pd.isnull(molecular_weight):
    return np.nan

  if units == 'nM':
    return (value * molecular_weight * 1e-6) / 1e6  # nM to mg/ml
  elif units == 'ug.mL-1':
    return value / 1000
  elif units == 'ug':
    return value / 1000
  elif units == "10'-11uM":
    return (value * molecular_weight * 1e-6) / 1e6
  elif units == "10'5uM":
    return (value * molecular_weight * 1e-6) / 1e6
  elif units == '%':
    return np.nan
  else:
    return np.nan


ic50_df['Standard Value'] = ic50_df.apply(convert_to_mg_ml, axis=1)
cc50_df['Standard Value'] = cc50_df.apply(convert_to_mg_ml, axis=1)
```
